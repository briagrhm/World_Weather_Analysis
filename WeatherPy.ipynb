```python
#Import Dependencies
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
from citipy import citipy
```

```python
#Create random latitutde and longtitudes combinations 

lat=np.random.uniform(-90,90, size=1500)
lngs=np.random.uniform(-180,180, size=1500)
lat_lngs = zip(lat,lngs)
lat_lngs
```

```python
coordinates = list(lat_lngs)
```

```python
for coordinate in coordinates:
    print(coordinate[0],coordinate[1])
```

```python
#Create a list for holding cities 
cities = []
#Identify city for each nearest latitude and longitude 
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    #If city is unique we will add it to the cities list 
    if city not in cities:
        cities.append(city)
#Print the city count 
len(cities)
```
